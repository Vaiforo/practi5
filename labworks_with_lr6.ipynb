Проверим KL-дивергенцию.

```python
import numpy as np

def get_distribution(series, allowed_values):
    counts = series.value_counts().reindex(allowed_values, fill_value=0)
    probs = counts / counts.sum()
    return probs.values
```

```python
col = 'Скорость перехода через портал'

allowed_values = (
    X_train[col]
    .dropna()
    .sort_values()
    .unique()
)

p_original = get_distribution(
    X_train[col].dropna(),
    allowed_values
)
```

```python
p_median = get_distribution(
    X_train_median['Скорость перехода через портал'],
    allowed_values
)
```

```python
p_moda = get_distribution(
    X_train_moda['Скорость перехода через портал'],
    allowed_values
)
```

```python
def iqr_outliers_mask(s: pd.Series, k: float = 1.5) -> pd.Series:
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    low = q1 - k * iqr
    high = q3 + k * iqr
    return (s < low) | (s > high)


def robust_z_outliers_mask(s: pd.Series, thr: float = 3.5) -> pd.Series:
    med = s.median()
    mad = (s - med).abs().median()
    if mad == 0:
        return pd.Series(False, index=s.index)
    rz = 0.6745 * (s - med) / mad
    return rz.abs() > thr
```

```python
potencial_cols = ["Печать Чародея"]

outlier_idx_by_col = {}
outlier_idx_all = set()

for col in potencial_cols:
    s = X_train[col]

    mask_iqr = iqr_outliers_mask(s, k=1.5)
    mask_rz = robust_z_outliers_mask(s, thr=3.5)

    print(f"{col}: IQR={len(set(s.index[mask_iqr]))} RZ={len(set(s.index[mask_rz]))}")

    mask_both = mask_iqr & mask_rz
    idx_col = set(s.index[mask_both])

    outlier_idx_by_col[col] = idx_col
    outlier_idx_all |= idx_col

len(outlier_idx_all)
```

- интерпретируем лучшую модель (важности / permutation importance).



```python
# === ЛР6: отбор моделей/метрик, обучение, интерпретация ===
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler, RobustScaler

from sklearn.linear_model import Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.inspection import permutation_importance

import optuna

SEED = 42
np.random.seed(SEED)

# -----------------------------
# 0) Проверяем, что нужные данные есть
# -----------------------------
needed = ["X_train", "X_test", "y_train", "y_test"]
missing = [v for v in needed if v not in globals()]
if missing:
    raise RuntimeError(f"Перед ЛР6 должны быть определены: {needed}. Не найдены: {missing}")

X_train_base = X_train.copy()
X_test_base  = X_test.copy()

# -----------------------------
# 1) Версии данных (как в ЛР5: без скалера и с 3 скалерами)
#    + используем версию с новыми признаками (X_new), но строим ее корректно и для train, и для test.
# -----------------------------
def make_features(df: pd.DataFrame) -> pd.DataFrame:
    # Генерация признаков из ЛР5 (делаем безопасно: errors="ignore")
    df = df.copy()

    def safe_get(col):
        return df[col] if col in df.columns else 0

    df["Общая мощность потоков"] = (
        safe_get("Сила Левого Потока") +
        safe_get("Сила Правого Потока") +
        safe_get("Приток Силы Потока")
    )

    df["Мощность ядра"] = (
        safe_get("Ритм магического ядра") *
        safe_get("Приток Силы Потока")
    )

    df["Общее давление на выходе"] = (
        safe_get("Приток давления Выдоха Истока") +
        safe_get("Давление выдоха Истока")
    )

    denom = safe_get("Эмульсия Истока")
    denom = denom.replace(0, np.nan) if isinstance(denom, pd.Series) else denom

    df["Магическая производительность"] = safe_get("Скорость перехода через портал") / denom
    df["Эффективность ядра"] = df["Мощность ядра"] / denom

    df["Магическая мощность"] = (
        safe_get("Эмульсия Истока") +
        df["Общее давление на выходе"] +
        safe_get("Пламя Стихий")
    )

    df.drop(
        columns=[
            "Сила Левого Потока",
            "Сила Правого Потока",
            "Приток Силы Потока",
            "Приток давления Выдоха Истока",
            "Давление выдоха Истока",
            "Эмульсия Истока",
            "Пламя Стихий",
            "Скорость перехода через портал",
        ],
        inplace=True,
        errors="ignore"
    )
    return df

X_train_feat = make_features(X_train_base)
X_test_feat  = make_features(X_test_base)

DATA_VERSIONS = {
    "base": (X_train_base, X_test_base),
    "feat": (X_train_feat, X_test_feat),
}

# -----------------------------
# 2) Визуальный анализ зависимостей (scatter) + "вектор корреляций"
# -----------------------------
def corr_vector_heatmap(X: pd.DataFrame, y: pd.Series, title: str, top_n: int = 25):
    X_num = X.select_dtypes(include=[np.number]).copy()
    corr = X_num.apply(lambda col: col.corr(y))
    corr_df = corr.to_frame("corr_to_target").sort_values("corr_to_target", key=lambda s: s.abs(), ascending=False)
    corr_df = corr_df.head(top_n)

    plt.figure(figsize=(6, max(6, 0.25 * len(corr_df))))
    sns.heatmap(corr_df, annot=True, fmt=".3f", center=0)
    plt.title(title)
    plt.show()

    return corr_df

def scatter_top_correlated(X: pd.DataFrame, y: pd.Series, corr_df: pd.DataFrame, top_n: int = 6):
    top_features = corr_df.index[:top_n].tolist()
    for col in top_features:
        plt.figure(figsize=(6, 4))
        sns.scatterplot(x=X[col], y=y, s=15, alpha=0.5)
        plt.title(f"{col} → целевой")
        plt.xlabel(col)
        plt.ylabel("Гармония Бессмертия")
        plt.show()

corr_df_feat = corr_vector_heatmap(
    X_train_feat, y_train,
    title="Вектор корреляций (Pearson): признак ↔ Гармония Бессмертия (версия feat)"
)
scatter_top_correlated(X_train_feat, y_train, corr_df_feat, top_n=8)

# -----------------------------
# 3) Выбор пула моделей и метрик
# -----------------------------
# По лекции: для регрессии адекватно смотреть группу метрик (RMSE, MAE, R2, MAPE/SMAPE),
# а для подбора гиперпараметров выбрать одну целевую метрику. fileciteturn1file0

def regression_metrics(y_true, y_pred) -> dict:
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    rmse = mean_squared_error(y_true, y_pred, squared=False)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    mask = y_true != 0
    mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100 if mask.any() else np.nan

    return {"RMSE": rmse, "MAE": mae, "R2": r2, "MAPE_%": mape}

# Целевая метрика Optuna: RMSE (штрафует большие ошибки и в единицах целевого)
OPTUNA_SCORING = "neg_root_mean_squared_error"

# Пул моделей (≤ 5)
MODEL_POOL = {
    "Ridge": Ridge(random_state=SEED),
    "ElasticNet": ElasticNet(random_state=SEED, max_iter=20000),
    "RandomForest": RandomForestRegressor(random_state=SEED, n_jobs=-1),
    "HistGB": HistGradientBoostingRegressor(random_state=SEED),
    "KNN": KNeighborsRegressor(),
}

# -----------------------------
# 4) Preprocess (impute + onehot + scaler)
# -----------------------------
def make_preprocess(X: pd.DataFrame, scaler):
    num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = [c for c in X.columns if c not in num_cols]

    num_steps = [("imputer", SimpleImputer(strategy="median"))]
    if scaler is not None:
        num_steps.append(("scaler", scaler))
    num_pipe = Pipeline(num_steps)

    if len(cat_cols) > 0:
        try:
            ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
        except TypeError:
            ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)

        cat_pipe = Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("ohe", ohe),
        ])
        preprocess = ColumnTransformer([
            ("num", num_pipe, num_cols),
            ("cat", cat_pipe, cat_cols),
        ])
    else:
        preprocess = ColumnTransformer([
            ("num", num_pipe, num_cols),
        ])

    return preprocess

SCALERS = {
    "noscale": None,
    "standard": StandardScaler(),
    "minmax": MinMaxScaler(),
    "robust": RobustScaler(),
}

# -----------------------------
# 5) Search-space гиперпараметров для Optuna (словари/правила)
# -----------------------------
def suggest_params(trial: optuna.Trial, model_name: str) -> dict:
    if model_name == "Ridge":
        return {
            "alpha": trial.suggest_float("alpha", 1e-3, 1e3, log=True),
            "fit_intercept": trial.suggest_categorical("fit_intercept", [True, False]),
        }

    if model_name == "ElasticNet":
        return {
            "alpha": trial.suggest_float("alpha", 1e-4, 10.0, log=True),
            "l1_ratio": trial.suggest_float("l1_ratio", 0.0, 1.0),
            "fit_intercept": trial.suggest_categorical("fit_intercept", [True, False]),
        }

    if model_name == "RandomForest":
        return {
            "n_estimators": trial.suggest_int("n_estimators", 200, 800),
            "max_depth": trial.suggest_int("max_depth", 3, 30),
            "min_samples_split": trial.suggest_int("min_samples_split", 2, 20),
            "min_samples_leaf": trial.suggest_int("min_samples_leaf", 1, 10),
            "max_features": trial.suggest_categorical("max_features", ["sqrt", "log2"]),
            "bootstrap": trial.suggest_categorical("bootstrap", [True, False]),
        }

    if model_name == "HistGB":
        return {
            "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
            "max_depth": trial.suggest_int("max_depth", 2, 10),
            "max_leaf_nodes": trial.suggest_int("max_leaf_nodes", 15, 255),
            "min_samples_leaf": trial.suggest_int("min_samples_leaf", 10, 200),
            "l2_regularization": trial.suggest_float("l2_regularization", 1e-6, 1.0, log=True),
        }

    if model_name == "KNN":
        return {
            "n_neighbors": trial.suggest_int("n_neighbors", 2, 60),
            "weights": trial.suggest_categorical("weights", ["uniform", "distance"]),
            "p": trial.suggest_categorical("p", [1, 2]),
            "leaf_size": trial.suggest_int("leaf_size", 10, 60),
        }

    raise ValueError(f"Unknown model: {model_name}")

def build_estimator(model_name: str, params: dict):
    if model_name == "Ridge":
        return Ridge(random_state=SEED, **params)
    if model_name == "ElasticNet":
        return ElasticNet(random_state=SEED, max_iter=20000, **params)
    if model_name == "RandomForest":
        return RandomForestRegressor(random_state=SEED, n_jobs=-1, **params)
    if model_name == "HistGB":
        return HistGradientBoostingRegressor(random_state=SEED, **params)
    if model_name == "KNN":
        return KNeighborsRegressor(**params)
    raise ValueError(f"Unknown model: {model_name}")

# -----------------------------
# 6) Optuna + k-Fold CV обучение
# -----------------------------
cv = KFold(n_splits=5, shuffle=True, random_state=SEED)

# Чтобы ноутбук был «адекватно быстрый». Если нужно — увеличьте.
N_TRIALS = {
    "Ridge": 30,
    "ElasticNet": 40,
    "RandomForest": 25,
    "HistGB": 35,
    "KNN": 25,
}

def objective(trial, model_name: str, X_tr: pd.DataFrame, y_tr: pd.Series, scaler):
    params = suggest_params(trial, model_name)
    estimator = build_estimator(model_name, params)

    preprocess = make_preprocess(X_tr, scaler)
    pipe = Pipeline([
        ("preprocess", preprocess),
        ("model", estimator),
    ])

    scores = cross_val_score(
        pipe,
        X_tr,
        y_tr,
        cv=cv,
        scoring=OPTUNA_SCORING,
        n_jobs=-1
    )
    rmse = -scores.mean()
    return rmse

records = []
best_pipelines = {}

for data_name, (X_tr, X_te) in DATA_VERSIONS.items():
    for scaler_name, scaler in SCALERS.items():
        version_name = f"{data_name}__{scaler_name}"

        for model_name in MODEL_POOL.keys():
            study = optuna.create_study(
                direction="minimize",
                sampler=optuna.samplers.TPESampler(seed=SEED)
            )
            study.optimize(
                lambda trial: objective(trial, model_name, X_tr, y_train, scaler),
                n_trials=N_TRIALS[model_name],
                show_progress_bar=False
            )

            best_params = study.best_params
            best_cv_rmse = study.best_value

            preprocess = make_preprocess(X_tr, scaler)
            final_pipe = Pipeline([
                ("preprocess", preprocess),
                ("model", build_estimator(model_name, best_params)),
            ])
            final_pipe.fit(X_tr, y_train)

            pred_train = final_pipe.predict(X_tr)
            pred_test = final_pipe.predict(X_te)

            m_train = regression_metrics(y_train, pred_train)
            m_test = regression_metrics(y_test, pred_test)

            rec = {
                "data_version": version_name,
                "model": model_name,
                "cv_RMSE": best_cv_rmse,
                **{f"train_{k}": v for k, v in m_train.items()},
                **{f"test_{k}": v for k, v in m_test.items()},
                "best_params": best_params,
            }
            records.append(rec)
            best_pipelines[(version_name, model_name)] = final_pipe

df_results = pd.DataFrame(records).sort_values(["test_RMSE", "cv_RMSE"]).reset_index(drop=True)

print("Топ-10 по test_RMSE:")
display(df_results[[
    "data_version", "model", "cv_RMSE",
    "test_RMSE", "test_MAE", "test_R2", "test_MAPE_%"
]].head(10))

best_row = df_results.iloc[0]
best_key = (best_row["data_version"], best_row["model"])
best_model = best_pipelines[best_key]

print("\nЛучшая модель:")
print(best_row[["data_version", "model", "cv_RMSE", "test_RMSE", "test_MAE", "test_R2"]])
print("\nBest params:", best_row["best_params"])

# -----------------------------
# 7) Интерпретация лучшей модели
# -----------------------------
best_data_tag = best_row["data_version"].split("__")[0]  # base / feat
X_tr_best, X_te_best = DATA_VERSIONS[best_data_tag]

# 7.1 Permutation importance — подходит для любых моделей
perm = permutation_importance(
    best_model,
    X_te_best,
    y_test,
    n_repeats=10,
    random_state=SEED,
    n_jobs=-1,
    scoring=OPTUNA_SCORING
)

preprocess = best_model.named_steps["preprocess"]
try:
    feat_names = preprocess.get_feature_names_out()
except Exception:
    feat_names = np.array([f"f{i}" for i in range(len(perm.importances_mean))])

imp = pd.DataFrame({
    "feature": feat_names,
    "importance": perm.importances_mean
}).sort_values("importance", ascending=False)

plt.figure(figsize=(8, 6))
sns.barplot(data=imp.head(15), x="importance", y="feature")
plt.title("Permutation Importance (top-15) для лучшей модели")
plt.show()

# 7.2 Pred vs Actual + остатки
pred_test_best = best_model.predict(X_te_best)
residuals = y_test.values - pred_test_best

plt.figure(figsize=(6, 5))
sns.scatterplot(x=pred_test_best, y=y_test.values, s=15, alpha=0.6)
plt.title("Predicted vs Actual (test)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(residuals, bins=30, kde=True)
plt.title("Распределение остатков (test)")
plt.xlabel("y_true - y_pred")
plt.show()
```
